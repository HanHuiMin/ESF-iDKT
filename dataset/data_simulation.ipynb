```python
#生成随机模拟数据
#确定参数：学生数目M，题目数量N，概念数量K
import numpy as np
import random
import math
import csv
import torch
import torch.nn
#题目总数
question_num=10  

#题目难度上限
question_diff_sup=2.0 
#学生总数
student_num=500 
#遗忘参数
r=0.1
beta=0.1
#知识点个数
concept_num=5
#答题序列长度
answer_len=question_num*3
#生成数据集编号
index=0
diff_unit = question_diff_sup/question_num
```

```python
#首先，生成随机的Q-matrix（N*K），题目的各个维度之间无关，且题目难度与题目编号无关,同时，生成每道题目对学生能力提升的矩阵，与难度有关（N*K）
question_diff={}
question_learn={}
q_matrix = []
diff = []

for i in range(0,question_num):
    diff_list=[]
    learn_list=[]
    for k in range(concept_num):
#         difficulty=max(diff_unit*(1+random.uniform(0.1,0.4)*i)*np.random.normal(0.5,2),0)
        difficulty=max(diff_unit*(0.2+random.uniform(0.1,0.2)*i)+random.uniform(-diff_unit*k, diff_unit*k/2),0)
        if difficulty<0:
            difficulty=0
        diff_list.append(difficulty)
        learn_list.append(difficulty*random.uniform(0.1, 0.6))
    question_diff[i]=diff_list
    q_matrix.append(diff_list)
    question_learn[i]=learn_list
```

```python
#然后，随机生成学生的初始能力值（M*K）
stu_init = np.random.rand(student_num,concept_num)
for i in range(student_num):
    for k in range(concept_num):
#         stu_init[i][k] = random.uniform(0,diff_unit*6)
        stu_init[i][k] = max(np.random.normal(diff_unit*2,2),0)
# print(stu_init)
stu_state = np.zeros((student_num,answer_len,concept_num))
for i in range(student_num):
    stu_state[i][0]=stu_init[i]
```

```python
def gen_result_torch(question,student):
    vector = 0.0
    q = torch.FloatTensor(question)
    s = torch.FloatTensor(student)
    cosine = torch.cosine_similarity(q,s,dim=0)
#     projection = cosine.repeat(1,concept_num)
    projection = s*cosine
    vector = torch.norm(projection,dim=0)-torch.norm(q,dim=0)
    if vector>0:
        return 1,vector
    else:
        return 0,vector
```

```python
#动态生成学生答题序列 如果学生做对某题就不再做了，如果做不对有一定概率继续做
#所有题目的字典，用于标记题目是否答对
question_dict={}
for i in range(question_num):
    question_dict[i]=0
#为每一个学生生成答题序列
column1={} #学生答题序列长度
column2={} #学生答题编号序列
column3={} #学生答题结果序列
column4={} #学生能力差值序列（仅作为参考）
column5={} #学生答题时间序列
num_total=0
num_wrong=0
for i in range(student_num):
    #先得到初始化列表用于标记学生是否答对题目
    q_dict=question_dict.copy()
    length = np.random.randint(answer_len/3,answer_len)#为该学生生成一个答题序列的长度
    num_total+=length
    column1[i] = length
    q_list=[]
    result_list=[]
    time_list=[]
    len_seq=0
    cur_time=np.random.randint(0,30)
    for j in range(length):#生成答题编号
        #从目前未答对的题目列表中随机生成该时刻答题的题目编号
        cur_q_list = list(q_dict.keys())
        if len(cur_q_list)==0:
            break
        item = np.random.randint(0,len(cur_q_list))
        q_id = cur_q_list[item]
        q_list.append(q_id)
        #随机生成此次答题与上次答题的时间差
        delta_time = np.random.randint(2,7)
        cur_time += delta_time
        time_list.append(cur_time)
        #计算学生能否答对该题
        correct, vector = gen_result_torch(question_diff[q_id],stu_state[i][j])
        result_list.append(correct)
        #根据答题结果，更新学生能力，以及更新答题队列
        learn_vector = []
        if correct==1:
            q_dict.pop(q_id) #答对题目，将该题目编号删去
            for k in range(concept_num):
                learn_vector.append(question_diff[q_id][k]*random.uniform(0.2,0.5))
        else:
            num_wrong+=1
            for k in range(concept_num):
                learn_vector.append(question_diff[q_id][k]*random.uniform(0.1,0.2))
        #更新下一时刻学生状态
        for k in range(concept_num):
            stu_state[i][j+1][k] = max(np.random.normal((stu_state[i][j][k]+learn_vector[k])*math.exp(-beta-r*delta_time),0.5),0)
    column1[i]=len(q_list)
    column2[i]=q_list
    column3[i]=result_list
    column5[i]=time_list
```

```python
print("总答题数：",num_total)
print("错误数：",num_wrong)
print("正确数：",num_total-num_wrong)
```

```python
#输出为csv格式
folderpath = "s"+str(student_num)+"_q"+str(question_num)+"_kc"+str(concept_num)
f_train = open(folderpath+"/s"+str(student_num)+"_q"+str(question_num)+"_kc"+str(concept_num)+"_train.csv", "w", newline='')
write_train = csv.writer(f_train, dialect='excel')
f_test = open(folderpath+"/s"+str(student_num)+"_q"+str(question_num)+"_kc"+str(concept_num)+"_test.csv", "w", newline='')
write_test = csv.writer(f_test, dialect='excel')
write_train.writerow(["stu_id","question_id","result","time"])
write_test.writerow(["stu_id","question_id","result","time"])
for i in range(student_num):
    if i%5!=0:
        for j in range(column1[i]):
            write_train.writerow(["stu_"+str(i), str(column2[i][j]),str(column3[i][j]), str(column5[i][j])])
    else:
        for j in range(column1[i]):
            write_test.writerow(["stu_"+str(i), str(column2[i][j]),str(column3[i][j]), str(column5[i][j])])
```

```python
#输出学生能力状态
folderpath = "s"+str(student_num)+"_q"+str(question_num)+"_kc"+str(concept_num)
# np.save(folderpath+"/state"+str(index),stu_state)
np.save(folderpath+"/qm"+str(index),np.array(q_matrix))
#将答题结果输出
# csv_output_train = open("学生"+str(student_num)+"题目"+str(question_num)+"概念数"+str(concept_num)+"/答题序列数据_train"+str(index)+".csv",'w', newline='')
csv_output_train = open(folderpath+"/s"+str(student_num)+"_q"+str(question_num)+"_kc"+str(concept_num)+"_train"+str(index)+".csv",'w', newline='')
csv_write_train = csv.writer(csv_output_train,dialect='excel')
# csv_output_test = open("学生"+str(student_num)+"题目"+str(question_num)+"概念数"+str(concept_num)+"/答题序列数据_test"+str(index)+".csv",'w', newline='')
csv_output_test = open(folderpath+"/s"+str(student_num)+"_q"+str(question_num)+"_kc"+str(concept_num)+"_test"+str(index)+".csv",'w', newline='')

csv_write_test = csv.writer(csv_output_test,dialect='excel')

j=0
for i in range(student_num):
    if i%5!=0:
        csv_write_train.writerow([column1[i]])
        csv_write_train.writerow(column2[i])
        csv_write_train.writerow(column3[i]) 
    else:
#         print("j=",j)
#         print(column2[i])
        csv_write_test.writerow([column1[i]])
        csv_write_test.writerow(column2[i])
        csv_write_test.writerow(column3[i])   
        np.save(folderpath+"/state"+str(index)+'/'+str(j),stu_state[i])
        np.save(folderpath+"/state"+str(index)+'/答题序列'+str(j),column2[i])
        j=j+1
csv_output_train.close()
csv_output_test.close()
```

```python
csv_output = open(folderpath+"/学生能力数据"+str(index)+".csv",'w', newline='')
csv_write = csv.writer(csv_output,dialect='excel')
csvdata = ["student_id","timestep","able"]
csv_write.writerow(csvdata) 
for i in range(student_num):
    for t in range(answer_len):        
        csvdata = ["student_"+str(i),t,stu_state[i][t]]
#         print(csvdata)
        csv_write.writerow(csvdata)
csv_output.close()
```
