```python
from pyecharts.charts import Bar, Grid, Line,Scatter
from pyecharts import options as opts
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
step_list = [i for i in range(0,29)]
line = Line()
line.load_javascript()
l_origin_state = torch.sigmoid(origin_state)
l_origin_state = l_origin_state.numpy()
l_model_state = torch.sigmoid(model_state)
l_model_state = l_model_state.numpy()
l_DKT_state = torch.sigmoid(DKT_state)
l_DKT_state = l_DKT_state.numpy()
l_DKTp_state = torch.sigmoid(DKTp_state)
l_DKTp_state = l_DKTp_state.numpy()
```

```python
line.add_xaxis(step_list[1:len(step)])
for i in range(0,1):
    line.add_yaxis("原始学生知识点"+str(i),list(l_origin_state[:,i]),is_smooth=True)
    line.add_yaxis("m7学生知识点"+str(i),list(l_model_state[:,i]),is_smooth=True)
    line.add_yaxis("DKT学生知识点"+str(i),list(l_DKT_state[:,i]),is_smooth=True)
#     line.add_yaxis("DKT+学生知识点"+str(i),list(l_DKTp_state[:,i]),is_smooth=True)
line.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
line.render_notebook()
```

```python
#计算学生能力相似度
state_list=[]
state_list.append(torch.tensor(origin_state))
state_list.append(torch.tensor(DKT_state))
state_list.append(torch.tensor(DKTp_state))
state_list.append(torch.tensor(model_state))
smi_list=[]
smi_matrix=[]
len_seq = origin_state.shape[0]
print(len_seq)
one = torch.ones((len_seq))
print(one)
full = torch.norm(one)
print(full)
for i in range(len(state_list)):
    smi_list=[]
    for j in range(len(state_list)):
        smi = torch.cosine_similarity(state_list[i],state_list[j],1)
        smi = torch.norm(smi)
        smi = smi/full
#         smi = torch.sigmoid(smi)
        smi_list.append(smi.item())
    smi_matrix.append(smi_list)
print(len(smi_matrix))

smi_matrix = np.array(smi_matrix)
# smi_matrix = smi_matrix/3.16
print(smi_matrix)
```

```python
index_list = ['GT','DKT','DKT+','ESF-iDKT']
plt.figure(figsize=(15, 4))
f_plot_1 = sns.heatmap(smi_matrix,vmin = 0.0,vmax=1,square=True,cmap=plt.cm.Blues,xticklabels=index_list,yticklabels=index_list,annot=True,fmt=".3f",linewidths=.1)
plt.show()
figure = f_plot_1.get_figure()
figure.savefig('stu91_similarity.pdf', bbox_inches='tight')
```
